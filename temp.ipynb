```python
import plotly.graph_objs as go
import plotly.io as pio
from plotly.subplots import make_subplots

pio.renderers.default = "browser"
from scipy.io import wavfile
from sound_params import *

# read WAV file
framerate, sound = wavfile.read('./file_example_WAV_1MG.wav')
sound = sound[:, 0]
nframes = len(sound)
duration = nframes / framerate
times = np.array([i/framerate for i in range(nframes)])


def plot_wave(times, sound, title, xaxis,  frame_size = 8):
    return go.Scatter(x=times, y=sound)
    # layout = go.Layout(title=title, xaxis=dict(title=xaxis), yaxis=dict(title="Amplitude"))
    #
    # fig = go.Figure(data=[trace], layout=layout)
    # return fig
```

```python
frame_size = 8
fig = make_subplots(rows=3, cols=2)

fig.add_trace(
    go.Scatter(x=times, y=sound),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=times[frame_size//2:-frame_size//2], y=volume(sound, frame_size)),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(x=times[frame_size//2:-frame_size//2], y=zcr(sound, frame_size)),
    row=2, col=1
)

fig.update_layout(height=600, width=800, title_text="Side By Side Subplots")
fig.show()
```

```python
frame_size = 8
plot_wave(times[4:-4], volume(sound, window_size=8))
```
